```python
import cv2
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt

def process_image_with_paddleocr(image_path):
    """
    Process an image using PaddleOCR, display the text with bounding boxes,
    and sort the extracted text row and column-wise.

    :param image_path: Path to the input image.
    :return: List of sorted extracted text with bounding boxes.

    """

    ocr = PaddleOCR(use_angle_cls=True, lang='en', use_gpu=False)



    image = cv2.imread(image_path)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


    #  Run OCR
    results = ocr.ocr(gray)


    #  Draw bounding boxes and text on the image
    boxes = [line[0] for line in results[0]]
    texts = [line[1][0] for line in results[0]]
    scores = [line[1][1] for line in results[0]]

    # Draw the boxes on the image
    image_for_drawing = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
    annotated_image = draw_ocr(image_for_drawing, boxes, texts, scores, font_path='Humor-Sans.ttf')

    # Visualize the annotated image
    plt.figure(figsize=(10, 10))
    plt.imshow(annotated_image)
    plt.axis('off')
    plt.title("Text and Bounding Boxes")
    plt.show()

    # Step 4: Sort text row-wise and column-wise
    sorted_results = sorted(
        results[0], key=lambda x: (x[0][0][1], x[0][0][0])
    )  # Sort by Y-coordinate, then X-coordinate

    # Format the output for better readability
    formatted_results = [
        {"text": line[1][0], "confidence": line[1][1], "box": line[0]} for line in sorted_results
    ]

    # Print sorted results
    print("Sorted Text:")
    for item in formatted_results:
        print(f"Text: {item['text']}, Confidence: {item['confidence']:.2f}, Box: {item['box']}")

    return formatted_results

# Example usage
image_path = "/content/form3.1.jpeg"
sorted_text_data = process_image_with_paddleocr(image_path)
```

```python
import cv2
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt

def annotate_image_with_text(image_path):
    """
    Annotates an image with the extracted text directly on its bounding boxes.

    :param image_path: Path to the input image.
    :return: None (displays the annotated image).
    """
    # Initialize PaddleOCR
    ocr = PaddleOCR(use_angle_cls=True, lang='en', use_gpu=False)


    # Read the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Unable to load image. Check the path.")
        return

    # Convert to RGB for display
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    #  Perform OCR
    results = ocr.ocr(image_path)

    # Step 3: Draw text on bounding boxes
    for line in results[0]:
        # Extract bounding box, text, and confidence
        box = line[0]  # Bounding box
        text = line[1][0]  # Extracted text
        confidence = line[1][1]  # Confidence score

        # Convert coordinates to integers
        box = [[int(point[0]), int(point[1])] for point in box]

        # Draw bounding box
        cv2.polylines(image, [np.array(box)], isClosed=True, color=(0, 255, 0), thickness=2)

        # Put text above the bounding box
        text_position = (box[0][0], box[0][1] - 10)  # Slightly above the top-left corner of the box
        cv2.putText(image, text, text_position, cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=0.6, color=(255, 0, 0), thickness=1, lineType=cv2.LINE_AA)

    #  Display the annotated image
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Annotated Image with Text")
    plt.show()

# Example usage
image_path = "/content/form3.1.jpeg"
annotate_image_with_text(image_path)
```

```python
import cv2
from paddleocr import PaddleOCR

def sort_text_row_col(image_path):
    """
    Extracts text from an image and sorts it row-wise and column-wise.

    :param image_path: Path to the input image.
    :return: List of rows, each containing column-sorted text.
    """
    # Initialize PaddleOCR
    ocr = PaddleOCR(use_angle_cls=True, lang='en', use_gpu=False)


    # Perform OCR on the image
    results = ocr.ocr(image_path)

    # Extract bounding boxes and text
    extracted_data = []
    for line in results[0]:
        box = line[0]  # Bounding box
        text = line[1][0]  # Extracted text
        y_center = (box[0][1] + box[2][1]) / 2  # Calculate y-center for row grouping
        x_center = (box[0][0] + box[2][0]) / 2  # Calculate x-center for column sorting
        extracted_data.append((y_center, x_center, text))

    # Sort data by y-center (row-wise), then by x-center (column-wise)
    extracted_data.sort(key=lambda item: (item[0], item[1]))

    # Group data into rows
    row_tolerance = 15  # Tolerance for considering text in the same row
    rows = []
    current_row = []

    for i, data in enumerate(extracted_data):
        if i == 0 or abs(data[0] - extracted_data[i - 1][0]) < row_tolerance:
            current_row.append(data)
        else:
            rows.append(current_row)
            current_row = [data]
    if current_row:
        rows.append(current_row)

    # Sort text within each row by x-center and prepare final sorted output
    sorted_rows = []
    for row in rows:
        row.sort(key=lambda item: item[1])
        sorted_rows.append([item[2] for item in row])

    return sorted_rows

# Example usage
image_path = "/content/form3.1.jpeg"
sorted_text = sort_text_row_col(image_path)

# Display sorted rows and columns
for i, row in enumerate(sorted_text):
    print(f"Row {i + 1}: {row}")
```
